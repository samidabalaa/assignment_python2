```python
# THis is a  student management system 
def menu():
    print(" [[[[[   STUDENT   ]]]]]   [[[[[ MANAGEMENT ]]]]]    [[[[[[ SYSTEM ]]]]]]  \n")
    print(" [#####] [####] [####] M [####] E [####]  N [####] U [####] [####] ")
    print("\n \/\/\/\/ select /\/\/\/\/\ your \/\/\/\/\/ choice /\/\/\/\/\ ")
    print("\n0.REGISTER NEW STUDENT ")
    print("1.SEARCH FOR STUDENT ")
    print("2.UPDATE STUDENT NAME")
    print("3.UPDATE STUDENT GPA")
    print("4.DELETE STUDENT BY ID")
    print("5.COUNT AND DISPLAY TOTAL NUMBER OF STUDENTS ")
    print("6.COUNT TOTAL NUMBER OF MALE AND FEMALE STUDENTS ")
    print("7.DISPLAY NAME AND DEPARTMENT OF TOP SCORER  STUDENTS FOR EACH DEPARTMENT ")
    print("8.DISPLAY NAME AND DEPARTMENT OF TOP SCORER FEMALE  STUDENTS FOR EACH DEPARTMENT ")
    print("9.LIST NAMES OF STUDENT WHO SCORED GREATER THAN  A GIVEN GPA ")
    print("10.SHOW FREQUENT STUDENT NAMES ")
    print("11.SHOW ALL ")
    print("12.SHOW TOTAL NUMBER OF STUDENTS IN EACH DEPARTMENT")
    print("13.exit")
    try:
        choice=int(input("enter your choice : "))
        if choice==0:
            register_student()
        elif choice==1:
            search_student()
        elif choice==2:
            update_student_name()
        elif choice==3:
            update_student_gpa()
        elif choice==4:
            delete_student()
        elif choice==5:
            count_student()
        elif choice==6:
            count_gender()
        elif choice==7:
            top_scorer_department()
        elif choice==8:
            top_female_scorer_department()
        elif choice==9:
            gpa_threshold()
        elif choice==10:
            frequent_names()
        elif choice==11:
            show_all()    
        elif choice==12:
            students_per_department()
        elif choice==13:
            SystemExit
        else:
            print(" invalid choice")
    except ValueError:
            print("Invalid input; please enter a number (0-12).")
            menu()  
# Function to register student
def register_student():
    with open("student_file.txt", 'a+') as file:
        student_id = input("Enter student id: ")
        for line in file :
            if line.split(";")[0]==student_id:
                print(" error, student already exists ! ")
                print(" do you want to continue in registration page")
                choice=input("enter yes/no : ")
                if choice=="yes":
                    register_student()    
                else:
                    menu()
                return
            
    st_fname = input("Enter first name: ")
    #   while  statment to check wheather  the input  name is string or not
    while True:
        if st_fname.isalpha():
            break
        else:
            while True:
                print("invalid name , name must has to be string")
                print(" do you want to continue")
                choice=input("enter yes/no : ")
                if choice=="yes":
                    st_fname = input("Enter student name (please use only string ): ")
                    if st_fname.isalpha():
                        break
                else:
                    menu()
    st_lname = input("Enter last name: ")
    while True:
        if st_lname.isalpha():
            break
        else:
            while True:
                print("invalid name , name must has to be string")
                print(" do you want to continue")
                choice=input("enter yes/no : ")
                if choice=="yes":
                    st_lname = input("Enter student name (please use only string ): ")
                    if st_lname.isalpha():
                        break
                else:
                    menu()    






    st_gender=input("Enter you gender (f/m) :")
    while True:
        if (st_gender=='f' and len(st_gender)==1) or (st_gender=='m' and len(st_gender)==1):
            break
        else:
            print(" invalid input, your input has to be f or m ")
            print(" do you want to continue")
            choice=input("enter yes/no : ")
            if choice=="yes":
                st_gender=input("Enter you gender (f/m) :")
                if (st_gender=='f' and len(st_gender)==1) or (st_gender=='m' and len(st_gender)==1):
                    break
                else:
                    menu() 



    st_department=input("Enter your department :")
    st_gpa =input("Enter student GPA: ")
    with open("student_file.txt", "a") as file:
        file.write(f"{ student_id};{st_fname};{st_lname};{st_gender};{st_department};{st_gpa}\n")
    print("student registered succesfully !")
    print(" do you want to continue in registration page")
    choice=input("enter yes/no : ")
    if choice=="yes":
        register_student()    
    else:
        menu() 
# Function to search for a student by ID
def search_student():
    found = False
    id_serached = input("Enter student ID to search: ")
    with open("student_file.txt", 'r') as file:
        for line in file:
            try:
                student_id,st_fname,st_lname,st_gender,st_department,st_gpa=line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if student_id.strip() == id_serached.strip():
                print("############# Data found ###################")
                print(f" first Name:{st_fname},last Name:{st_fname},gender:{st_gender},department:{st_department}, GPA: {st_gpa}")
                found = True
                break
    if not found:
        print("Student does not exist")
    menu()
# Function to update student name by using Id
def  update_student_name():
    s_id= input("Enter student ID: ")
    updated = False
    lines = []
    with open("student_file.txt", 'r') as file:
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname,st_gender,st_department,st_gpa = line.split(';')
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if s_id.strip() == student_id.strip():
                new_fname = input("Enter corrected first  name: ").strip()
                

                while True:
                    if new_fname.isalpha():
                        break
                    else:
                        while True:
                            print("invalid name , name must has to be string")
                            print(" do you want to continue")
                            choice=input("enter yes/no : ")
                            if choice=="yes":
                                new_fname = input("Enter student name (please use only string ): ")
                                if new_fname.isalpha():
                                    break
                            else:
                                menu()
                new_lname = input("Enter corrected last  name: ").strip()

                while True:
                    if new_fname.isalpha():
                        break
                    else:
                        while True:
                            print("invalid name , name must has to be string")
                            print(" do you want to continue")
                            choice=input("enter yes/no : ")
                            if choice=="yes":
                                new_fname = input("Enter student name (please use only string ): ")
                                if new_fname.isalpha():
                                    break
                            else:
                                menu()




                line = f"{s_id};{new_fname};{new_lname};{st_gender};{st_department};{st_gpa}\n"
                updated = True
                print(f"Data of student {new_fname}  {new_lname} updated successfully!")
            lines.append(line +'\n')
    if not updated:
        print(f"Student with ID {s_id} not found.")
    else:
        with open("student_file.txt", 'w') as file:
            file.writelines(lines)
    menu()
#  a function that  updates the student by using student id
def  update_student_gpa():
    st_id = input("Enter student ID: ")
    updated = False
    lines = []
    with open("student_file.txt", 'r') as file:
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname,st_gender,st_department,st_gpa = line.split(';')
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if st_id.strip() == student_id.strip():
                try:
                    new_gpa = float(input("Enter corrected GPA: "))
                    line = f"{student_id};{st_fname};{st_lname};{st_gender};{st_department};{new_gpa}\n"
                    updated = True
                    print(f"Data of student {st_fname} {st_lname} updated successfully!")
                except ValueError:
                    print("Invalid GPA entered. Please enter a numeric value.")
            lines.append(line + '\n')
    if not updated:
        print(f"Student with ID {st_id} not found.")
    else:
        with open("student_file.txt", 'w') as file:
            file.writelines(lines)
    menu()
#  the function that deletes the student data by using id
def delete_student():
    st_id = input("Enter student ID: ")
    deleted = False
    lines = []
    with open("student_file.txt", 'r') as file:
        for line in file:
            try:
                student_id,st_fname,st_lname,st_gender,st_department,st_gpa=line.split(";")
                
                
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if st_id != student_id:
                line=f"{student_id};{st_fname};{st_lname};{st_gender};{st_department};{st_gpa}"
                lines.append(line)
            else:
                deleted=True
                print(" student data deleted succefully ! ")
    if not deleted:
        print(f"Student with ID {st_id} not found.")
    else:
        with open("student_file.txt", 'w') as file:
            file.writelines(lines)
    menu()      
#  a function that counts total number of student
def count_student():
    count=0
    with open("student_file.txt","r") as file:
        for line in file:
            count+=1
    print(f"Total number of students: {count}")
    menu()
# a function that counts students based on gender
def count_gender():
    gender_counts = {'m': 0, 'f': 0}  
    with open("student_file.txt", "r") as file:
        for line in file:
            line = line.strip()
            try:
                student_id,st_fname,st_lname,st_gender,st_department,st_gpa=line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            gender_key = st_gender.strip().lower()
            if gender_key in gender_counts.keys():
                gender_counts[gender_key] += 1
            else:
                print(f"Unexpected gender value: {st_gender}")

    print(f"Total number of male students: {gender_counts['m']}")
    print(f"Total number of female students: {gender_counts['f']}")
    menu()
# a function that defines  top scorers of the department
def top_scorer_department():
    with open("student_file.txt", "r") as file:
        s_department = input("Enter the department: ")
        maximum_gpa = 0
        maximum_name = ""
        for line in file:
            line = line.strip()
            try:
                student_id,st_fname,st_lname,st_gender,st_department,st_gpa=line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if s_department == st_department and float(st_gpa) > maximum_gpa:
                maximum_gpa = float(st_gpa)
                maximum_fname = st_fname
                maximum_lname = st_lname
        print(f"Top scoring student in {st_department} department: {maximum_fname} {maximum_lname} with GPA {maximum_gpa}") 
    menu()
# a function that identifies top scorer female students in each department
def top_female_scorer_department():
    with open("student_file.txt", "r") as file:
        s_department = input("Enter the department: ")
        maximum_gpa = 0
        maximum_fname = ""
        maximum_lname = ""
        for line in file:
            line = line.strip()
            try:
                student_id,st_fname,st_lname,st_gender,st_department,st_gpa=line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if s_department == st_department and st_gender.lower() == "f" and float(st_gpa) > maximum_gpa:
                maximum_gpa = float(st_gpa)
                maximum_fname = st_fname
                maximum_lname = st_lname
        print(f"Top scoring female  student in {st_department} department: {maximum_fname}{maximum_lname} with GPA {maximum_gpa}") 
    menu()
# a function that  identifies students with greater than the given (accepted from the user) score
def gpa_threshold():
    s_gpa = float(input("Enter GPA threshold: "))
    with open("student_file.txt", "r") as file:
        print(f"Students with GPA greater than {s_gpa}:")
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname, st_gender, st_department, st_gpa = line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            if float(st_gpa) > s_gpa:
                print(f"Name: {st_fname} {st_lname}, GPA: {st_gpa}")

    menu()
#  the function that  identifies the frequent student name in the file
def frequent_names():
    name_count = {}
    with open("student_file.txt", "r") as file:
        print("** frequent ** names * among ** students ")
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname,st_gender, st_department, st_gpa = line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            name_key = st_fname.strip()
            if name_key in name_count:
                name_count[name_key] += 1
            else:
                name_count[name_key] = 1
            
    for name, count in name_count.items():
        if count > 1:
            print(f"    {name} :(occurs {count} times)")

    menu()
# a function that displays   all students registerd in file 
def show_all():
    print(" [###] [#####] [#####] [######] Student [####] file [#####] [######] [######] ")
    with open("student_file.txt", 'r') as file:
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname, st_gender, st_department, st_gpa = line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            print(f" ID: {student_id}\n first Name: {st_fname}\n last Name: {st_lname}\n Gender: {st_gender}\n Department: {st_department}\n GPA: {st_gpa}\n")
            print("[#####################]        [####################]      [####################]")
    menu()
#  the function that tells the number of students in each department
def students_per_department():
    department_counts = {}
    with open("student_file.txt", "r") as file:
        for line in file:
            line = line.strip()
            try:
                student_id, st_fname,st_lname, st_gender, st_department, st_gpa = line.split(";")
            except ValueError:
                print(f"Error processing line: {line}")
                continue
            department_key = st_department.strip()
            if department_key in department_counts:
                department_counts[department_key] += 1
            else:
                department_counts[department_key] = 1

    for department, count in department_counts.items():
        print(f"Total number of students in {department} department: {count}")

    menu()

menu()
```
